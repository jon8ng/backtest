```python
import yfinance as yf
import datetime as dt
import pandas as pd
```

```python
# Create a watchlist

watchlist = ['aapl',
             'tsla',]
```

```python
class options_filter:
    
    def __init__ (self):
        return
    
    def ticker_data(watchlist):
        td = []
        for i, d in enumerate(watchlist):
            td.append(yf.Ticker(d))       
        return td

    def get_options_expiry(ticker_data):
        oe = []
        for i, d in enumerate(ticker_data):
            oe.append(d.options)
        return oe
    
    def get_option_chain(option_expiry):
        oc = []
        for i, d in enumerate(option_expiry):          
            for x, y in enumerate(d):
                diff = dt.datetime.strptime(y,'%Y-%m-%d') - dt.datetime.today()
                if diff.days > 3:
                    if diff.days < 9:
                        oc.append(y)
        return oc
    
    def get_contracts(option_chain, tickers):
        cc = []
        pc = []
        filtc = []
        filtp = []
        
        for i, d in enumerate(option_chain): 
            cc.append(tickers[i].option_chain(date = d).calls)
            pc.append(tickers[i].option_chain(date = d).puts)
        
        for i, d in enumerate(cc):
            d['vol/oi'] = d['volume'].divide(d['openInterest'])
            
            fc = d.loc[(d['impliedVolatility'] <= 0.40) &
                          (d['inTheMoney'] == False) &
                          (d['vol/oi'] > 0.75) &
                          (d['volume'] > 200)
                         ]
            fc = fc[['contractSymbol',
                   'strike',
                   'lastPrice',
                   'volume',
                   'openInterest',
                   'impliedVolatility',
                   'vol/oi',
                  ]]
            fc = fc.sort_values(by='volume', ascending=False)
            filtc.append(fc)
            
        for i, d in enumerate(pc):
            d['vol/oi'] = d['volume'].divide(d['openInterest'])
            
            fp = d.loc[(d['impliedVolatility'] <= 0.40) &
                          (d['inTheMoney'] == False) &
                          (d['vol/oi'] > 0.75) &
                          (d['volume'] > 200)
                         ]
            fp = fp[['contractSymbol',
                   'strike',
                   'lastPrice',
                   'volume',
                   'openInterest',
                   'impliedVolatility',
                   'vol/oi',
                  ]]
            fp = fp.sort_values(by='volume', ascending=False)
            filtp.append(fp)

        return filtc, filtp
```
