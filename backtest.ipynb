```python
import yfinance as yf
import datetime as dt
```

```python
# Create a watchlist

watchlist = ['aapl',
             'tsla',]
```

```python
class options_filter:
    
    def __init__ (self):
        return
    
    def ticker_data(watchlist):
        td = []
        for i, d in enumerate(watchlist):
            td.append(yf.Ticker(d))       
        return td

    def get_options_expiry(ticker_data):
        oe = []
        for i, d in enumerate(ticker_data):
            oe.append(d.options)
        return oe
    
    def get_option_chain(option_expiry):
        oc = []
        for i, d in enumerate(option_expiry):          
            for x, y in enumerate(d):
                diff = dt.datetime.strptime(y,'%Y-%m-%d') - dt.datetime.today()
                if diff.days > 3:
                    if diff.days < 9:
                        oc.append(y)
        return oc
```
